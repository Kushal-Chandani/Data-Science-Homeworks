```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
da = pd.read_csv('EmployeeSalaryRegression.csv')
```

```python
#Question 1
```

```python
# Assuming you have a DataFrame 'da' with the relevant data

X = da['TotalWorkingYears']
y = da['MonthlyIncome']

# Fit the linear regression model
X = sm.add_constant(X)  # Add a constant (intercept) term
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```

```python
# Assuming you have 'da' and 'model' from the previous code

# Original data
plt.plot(da['TotalWorkingYears'], da['MonthlyIncome'], 'o', label='original data')

# Fitted values
X = da['TotalWorkingYears']
X = sm.add_constant(X)  # Add a constant (intercept) term
fitted_values = model.predict(X)

plt.plot(da['TotalWorkingYears'], fitted_values, 'r--.', label='fitted values')

plt.legend()
plt.show()
```

```python
# A p-value below the threshold of 0.05 signifies a statistically significant positive linear relationship
# between TotalWorkingYears and MonthlyIncome. In simpler terms, as an individual's TotalWorkingYears increase,
# their MonthlyIncome tends to rise accordingly. The strength of this relationship intensifies as the p-value decreases,
# and this robust connection is visually evident when inspecting the graph where data points closely track the regression line.
# In summary, our data provides compelling evidence for a strong positive correlation, and we can express this relationship
# through the linear equation:

# Let 'x' represent TotalWorkingYears. Let 'y' denote MonthlyIncome. The equation takes the form y = 1228 + 468x.

# In practical terms, this implies that for each additional year an employee works, their monthly income is expected
# to increase by $468. Additionally, the R-squared value of 0.597 reveals that approximately 77% of the variability in
# MonthlyIncome can be accounted for by variations in TotalWorkingYears. This high value indicates that the model fits
# the data well, which aligns with the common observation in most global companies: the longer an individual remains
# with the same employer, the more likely they are to see their salary increase due to the expertise and experience
# they accumulate over time. This phenomenon is consistently reflected in our dataset, as an employee's MonthlyIncome
# tends to rise as they accumulate more years of service.
```

```python
# Assuming you have a DataFrame 'da' with the relevant data

X = da['Age']
y = da['DistanceFromHome']

# Fit the linear regression model
X = sm.add_constant(X) 
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
# Original data
plt.plot(da['Age'], da['DistanceFromHome'], 'o', label='original data')

# Fitted values
X = da['Age']
X = sm.add_constant(X) 
fitted_values = model.predict(X)

plt.plot(da['Age'], fitted_values, 'r--.', label='fitted values')

plt.legend()
plt.show()
```

```python
# Looking at the table, we observe an R-squared value of 0.000. This value indicates that the variance in MonthlyIncome
# cannot be elucidated by changes in YearsAtCompany. This lack of explanatory power is also visually evident in the scatterplot,
# where the data points are dispersed and fail to conform to the regression line.

# Consequently, it is apparent that there is no discernible linear connection between YearsAtCompany and MonthlyIncome. Given
# this absence of a linear relationship, conducting an analysis of the p-value or other variables becomes superfluous.
```

```python
categorical_columns = da.select_dtypes(exclude=['int', 'float']).columns.tolist()
da_encoded = pd.get_dummies(da, columns=categorical_columns)

X = da_encoded.drop(['MonthlyIncome'], axis=1)
Y = da_encoded['MonthlyIncome']

X_with_const = sm.add_constant(X)  
model = sm.OLS(Y, X_with_const).fit()

model_summary = model.summary()
print(model_summary)
```

```python
# In the MonthlyIncome regression model, we determine the significance of input factors by analyzing the p-values associated with MonthlyIncome and all input variables.
# A p-value below 0.05 indicates a strong relationship, while a higher p-value suggests a weaker one.

# Least Influential Factors:
# Conversely, to identify the least significant inputs, we consider the three variables with the highest p-values: PerformanceRating, YearsInCurrentRole, and WorkLifeBalance.
# Their high p-values indicate minimal impact on MonthlyIncome calculations for employees.

# Most Influential Factors:
# To identify the most significant inputs, we focus on the three variables with the lowest p-values: JobLevel, TotalWorkingYears, and JobRole (including its subfields after one-hot encoding).
# These variables have the lowest p-values (0.0), signifying their substantial influence on MonthlyIncome calculations for employees.
```

```python
new_input = pd.DataFrame({
    'Age': [30],
    'Department': ['Research & Development'],
    'DistanceFromHome': [5],
    'Education': [3],
    'EducationField': ['Life Sciences'],
    'Gender': ['Male'],
    'HourlyRate': [50],
    'JobLevel': [2],
    'JobRole': ['Laboratory Technician'],
    'MaritalStatus': ['Single'],
    'NumCompaniesWorked': [2],
    'OverTime': ['No'],
    'PerformanceRating': [3],
    'StockOptionLevel': [1],
    'TotalWorkingYears': [10],
    'TrainingTimesLastYear': [3],
    'WorkLifeBalance': [3],
    'YearsAtCompany': [2],
    'YearsInCurrentRole': [2],
    'YearsSinceLastPromotion': [1],
    'YearsWithCurrManager': [2]
})

new_input = pd.concat([da, new_input], ignore_index=True)

new_input = new_input.drop(['MonthlyIncome'], axis=1)

new_input = pd.get_dummies(new_input, columns=[x for x in categorical_columns])

new_input = sm.add_constant(new_input)

new_input = new_input.tail(1)
predicted_income = model.predict(new_input)

print("The predicted MonthlyIncome is:", predicted_income.values[0])
```
