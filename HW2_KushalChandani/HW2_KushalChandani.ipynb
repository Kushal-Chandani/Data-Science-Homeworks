```python
#### Part 1 of Homework

#Question 2 Formatting

dfg = df.iloc[:, [14, 15, 16, 17]]
dfg.columns = dfg.iloc[0].astype(str)
dfg = dfg[1:]

columns_to_replace = ['Male', 'Female', 'NonBinary', 'Prefer not to answer']

for column in columns_to_replace:
    dfg[column] = dfg[column].replace('1', column)

dfg = dfg.fillna('')
dfg['Q2'] = dfg.apply(lambda row: ''.join(row.astype(str)), axis=1)
dfg = dfg[['Q2']]

dfg.head()
```

```python
# Question 8 Formatting

dfl = df.iloc[:, [23, 24, 25, 26]]
dfl.columns = dfl.iloc[0].astype(str)
dfl = dfl[1:]

columns_to_replace = ['Microsoft/Windows', 'Apple/MacBook', 'Other']

for column in columns_to_replace:
    dfl[column] = dfl[column].replace(['0', '1', '2'], column[:3])

dfl = dfl.fillna('')
dfl['Q8'] = dfl.apply(lambda row: ''.join(row.astype(str)), axis=1)
dfl = dfl[['Q8']]

dfl.head()
```

```python
# Question 9 Formatting

dff = df.iloc[:, [27, 28, 29]]
dff.columns = dff.iloc[0].astype(str)
dff = dff[1:]

columns_to_replace = ['Yes, Full Time', 'Working, but not Full Time', 'Not Working while attending Mason']

for column in columns_to_replace:
    if column == 'Yes, Full Time':
        dff[column] = dff[column].replace('1', 'full time')
    else:
        dff[column] = dff[column].replace(['0', '1', '2'], 'part time' if column == 'Working, but not Full Time' else 'not working')

dff = dff.fillna('')
dff['Q9'] = dff.apply(lambda row: ''.join(row.astype(str)), axis=1)
dff = dff[['Q9']]

dff.head()
```

```python
# Question 10 Formatting

dfc = df.iloc[:, [30, 31, 32, 33, 34]]
dfc.columns = dfc.iloc[0].astype(str)
dfc = dfc[1:]

columns_to_replace = ['Little/none', 'Some familiarity', 'Average user', 'Frequent use for projects', 'Fluent/expert']

for column in columns_to_replace:
    if column == 'Little/none':
        dfc[column] = dfc[column].replace('1', 'little/none')
    else:
        dfc[column] = dfc[column].replace(['0', '1', '2'], 'some' if column == 'Some familiarity' else 'average' if column == 'Average user' else 'frequent user' if column == 'Frequent use for projects' else 'expert')

dfc = dfc.fillna('')
dfc['Q10'] = dfc.apply(lambda row: ''.join(row.astype(str)), axis=1)
dfc = dfc[['Q10']]

dfc.head()
```

```python
# Question 11 Formatting

dfr = df.iloc[:, [35, 36, 37, 38, 39]]
dfr.columns = dfr.iloc[0].astype(str)
dfr = dfr[1:]

columns_to_replace = ['Little/none', 'Some familiarity', 'Average user', 'Frequent use for projects', 'Fluent/expert']

for column in columns_to_replace:
    if column == 'Little/none':
        dfr[column] = dfr[column].replace('1', 'little/none')
    else:
        dfr[column] = dfr[column].replace(['0', '1', '2'], 'some' if column == 'Some familiarity' else 'average' if column == 'Average user' else 'frequent user' if column == 'Frequent use for projects' else 'expert')

dfr = dfr.fillna('')
dfr['Q11'] = dfr.apply(lambda row: ''.join(row.astype(str)), axis=1)
dfr = dfr[['Q11']]

dfr.head()
```

```python
# Question 12 Formatting

dfp = df.iloc[:, [40, 41, 42, 43, 44]]
dfp.columns = dfp.iloc[0].astype(str)
dfp = dfp[1:]

columns_to_replace = ['Little/none', 'Some familiarity', 'Average user', 'Frequent use for projects', 'Fluent/expert']

for column in columns_to_replace:
    if column == 'Little/none':
        dfp[column] = dfp[column].replace('1', 'little/none')
    else:
        dfp[column] = dfp[column].replace(['0', '1', '2'], 'some' if column == 'Some familiarity' else 'average' if column == 'Average user' else 'frequent user' if column == 'Frequent use for projects' else 'expert')

dfp = dfp.fillna('')
dfp['Q12'] = dfp.apply(lambda row: ''.join(row.astype(str)), axis=1)
dfp = dfp[['Q12']]

dfp.head()
```

```python
#Part 2 of Homework

import numpy as np

da = pd.read_csv('formatted_580SurveyCleanupFinal.csv')
da.isnull().sum()
```

```python
#Dropping empty tables

da = da.drop(['Access Code', 'Email Address'], axis = 1)
```

```python
#replacing values

da['Q4'] = da['Q4'].replace({'7O': '70', '6l': '61', '5.2': '52'})
da['Q4'].fillna(da['Q4'].mode()[0], inplace=True)
```

```python
da.head()
```

```python
#filling columns

columns_to_fill = ['Q8', 'Q9', 'Q10', 'Q11', 'Q12']

for column in columns_to_fill:
    da[column].fillna(da[column].mode()[0], inplace=True)
```

```python
da.head()
```

```python
# filling missing values 

columns_to_fill = ['Region', 'City / Town / District']

for column in columns_to_fill:
    da[column].fillna(da[column].mode()[0], inplace=True)
```

```python
da.head()
```

```python
da['Q3'] = da['Q3'].str.replace(' ', '') #filling spaces with Q3
da['Q3'] = da['Q3'].replace('1993', '30') #replace 1993 with 30
da['Q3'].fillna(da['Q3'].mode()[0], inplace=True) #filling missing values with mode
```

```python
da.head()
```

```python
da['Q6'] = da['Q6'].str.title().str.strip().replace({
    'Bachelor of engineering ': 'Engineering',
    'Engineering ': 'Engineering',
    'Cse': 'Computer Science and Engineering',
    'Computer science engineering ': 'Computer Science and Engineering',
    'Computer engineering': 'Computer Science and Engineering',
    'B.tech in cs': 'Computer Science',
    'Criminology law and soci': 'Criminology law and social sciences',
    'Ece': 'Electrical and Computer Engineering',
    'Economic': 'Economics',
    'Yes': np.NaN,
})

da['Q6'].fillna(da['Q6'].mode()[0], inplace=True)
```

```python
da
```

```python
da['Q7'].fillna('Not responded', inplace=True)

da['Q8'] = da['Q8'].replace('Microsoft/Windows', 'Windows')
da['Q8'] = da['Q8'].replace('Apple/MacBook', 'Macbook')
da['Q8'] = da['Q8'].replace('Acer/Windows', 'Windows')

da['Q13'].fillna('Not responded', inplace=True)
```

```python
da.head()
```

```python
da['Q5'] = da['Q5'].str.capitalize()
da['Q5'] = da['Q5'].fillna('N/A', inplace=False)
da['Q5'] = da['Q5'].replace(['Usa', 'United states', 'United states ', 'Us'], 'USA')
da['Q5'] = da['Q5'].replace(['United states ; pakistan '], 'USA ; Pakistan')
```

```python
da.head()
```

```python
#Making CSV file

da.to_csv('cleaned_580SurveyCleanupFinal.csv', index = False)
```
